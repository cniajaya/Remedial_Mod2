There are 2 columns that would be nice if they are represented as numerical value rather than written; Num-of-Doors and Cylinders. For this 2 columns, replace the string as numerical value (Ex: if it is two then replace it with 2, etc.)

```python
def changenum(x):
    angka = ['one','two','three','four','five','six','seven','eight','nine','ten','eleven','twelve']
    if pd.isna(x):
        return x
    else:
        return (angka.index(x))+1
```

```python
auto['Num-of-Doors'] = auto['Num-of-Doors'].apply(changenum)
auto['Cylinders'] = auto['Cylinders'].apply(changenum)
```

Now convert these following columns into float object:
- Num-of-Doors

```python
auto['Num-of-Doors'] = auto['Num-of-Doors'].astype(float)
auto['Cylinders'] = auto['Cylinders'].astype(float)
auto['Bore'] = auto['Bore'].astype(float)
auto['Stroke'] = auto['Stroke'].astype(float)
auto['Horsepower'] = auto['Horsepower'].astype(float)
auto['Peak-rpm'] = auto['Peak-rpm'].astype(float)
```

```python
na = []
for i in auto.columns:
    for j,k in enumerate(auto[i]):
        if pd.isna(k):
            na.append(j)
na = [i for i in na]
na.sort()
auto.loc[na]
```

```python
# soal
```

Now we will try to fill the NaN value for some of the columns.
For the NaN in Num-of-Doors, Bore, and  Stroke columns fill the NaN value with the mean of their columns with the similar value in the Make and Body-Style columns. *e.g if you have NaN in num-of-doors with hatchback renault car then fill the NaN with the num-of-doors mean of the hatchback renault car. 
 

```python
for col in auto.columns:
    for i in na:
        if pd.isna(auto.loc[i,col]):
            auto.loc[i,col] = auto.groupby(by=['Body-Style','Make']).mean().loc[auto.loc[i,'Body-Style'],auto.loc[i,'Make']][col]
```

```python
from scipy.stats import shapiro, normaltest, anderson
def try_normal(data):
    result = {'Anderson' : {i:j for i,j in zip(anderson(data)[2], anderson(data)[1])}, 'Shapiro': shapiro(data)[1], 
     'K^2 Chi Square': normaltest(data)[1]}
    result['Anderson']['stat'] = anderson(data)[0]
    return result
```

```python
try_normal(auto['Price'])
```

In Anderson Test, stat value > critical value --> HO Rejected. It means data is not normal distributed

```python
from scipy.stats import mannwhitneyu
```

```python
result = list(mannwhitneyu(auto[auto['Make'] != 'jaguar']['Price'], 
                        auto[auto['Make'] == 'jaguar']['Price']))
print('U-Statistic:', result[0])
print('P-Value:', result[1])
```

From the statistic, it seems that there are significant differences between 
the price of the category make jaguar and not jaguar. (using significance level 5%)

```python
result = list(mannwhitneyu(auto[auto['Make'] != 'toyota']['Price'], 
                        auto[auto['Make'] == 'toyota']['Price']))
print('U-Statistic:', result[0])
print('P-Value:', result[1])
```

From the statistic, it seems that there are significant differences between 
the price of the category make toyota and not toyota.

```python
from pingouin import kruskal
kruskal(data=auto, dv='Price', between='Make')
```

Based on p-unc (or Uncorrected p-value) and using significance level 5%, there are relationship between make and price.
